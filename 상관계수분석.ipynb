```python
##################### 시도별 각 카테고리 개수 선 그래프
##################### 편차가 커 정규화 하여 선그래프 특징만 확인한다.
#################### 그래프의 흐름을 보니 각 카테고리별 개수끼리 서로 동행하는 모습도 보인다.
#################### 각 카테고리마다 다른 시도 보다 월등히 많은 것 을 보이는 위치가 다르게 나타났다.
plt.figure(figsize=(14,7))
ax = sns.lineplot(x='loc',y='숙박업소개수',data=df,color='r')
ax = sns.lineplot(x='loc',y='관광지개수',data=df,color='b')
ax = sns.lineplot(x='loc',y='액티비티개수',data=df,color='y')
ax = sns.lineplot(x='loc',y='음식점개수',data=df,color='g')
ax.set_title('전국 각 카테고리별 분포 선 그래프',
            pad=15.0)
ax.set_xlabel('지역')
ax.set_ylabel('정규화수치',rotation=0,labelpad=50,position=(0,0.45,0))
red = mpatches.Patch(color='red',label='숙박업소')
blue = mpatches.Patch(color='blue',label='관광지')
yellow = mpatches.Patch(color='yellow',label='엑티비티')
green = mpatches.Patch(color='green',label='음식점')
plt.legend(handles=[red,blue,yellow,green])
#plt.show()
plt.savefig('전국 각 카테고리별 분포 선그래프.png',bbox_inches='tight', pad_inches=0.5) 
```

```python
############# 관광지별 숙박시설,관광지,액티비티 개수간 상관관계 추후 음식점 추가 
corr = df.iloc[:,1:5].corr(method='pearson')
corr
####### 숙박시설은 타 카테고리와 모두 양의 상관관계를 보임
####### 관광지는 엑티비티 제외 양의 상관관계를 가짐
####### 액티비티는 숙박시설과 음식점과 양의 상관관계를 가지며 음식점과 상관관계는 약하다.
```

```python
########### 상관관계 heatmap
ax = plt.figure(figsize=(14,7))
ax = sns.heatmap(corr,cmap='YlGnBu',annot=True,fmt='0.2f',linewidths=.8,annot_kws={'size':20})
plt.title('각 카테고리별 개수간 상관관계',pad= 30)
#plt.show()
plt.savefig('각 카테고리별 개수간 상관관계 heatmap.png',bbox_inches='tight', pad_inches=0.5) 
```
