```python
import pandas as pd
import numpy as np
from konlpy.tag import Komoran
komoran=Komoran(max_heap_size = 1024)
from apyori import apriori
```

```python
df = pd.read_csv('연관분석임시/seoul.csv',encoding='utf8')
```

```python
df['0'] = df['0'].str.replace('[\W+]',' ')
```

```python
df = df.dropna()
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
contents =[]
for c in range(len(df)):
    temp = df['0'][c]
    contents.append(temp)
```

```python
len(contents)
```

```python
contents = []
for c in range(len(df)):
    try:
        temp = list(filter(lambda word : len(word)>1,komoran.nouns(df['0'][c])))
        contents.append(temp)
    except:
        print('에러발생'+str(c)+'번')
        continue
```

```python
len(contents)
```

```python
contents
```

```python
rules = apriori(contents,
               min_support=0.3,
               min_confidence=0.2)
result = list(rules)
```

```python
result_df = pd.DataFrame(None,columns=['lhs','rhs','support','confidence','lift'])
index=0
for row in result:
    support = row[1]
    order_st = row[2]
    for item in order_st:
        lhs = ','.join(x for x in item[0])
        rhs = ','.join(x for x in item[1])
        confidence = item[2]
        lift = item[3]
        if lift != 1:
            result_df.loc[index] =[lhs,rhs,support,confidence,lift]
            index+=1
```

```python
pd.options.display.max_rows = 644
result_df.loc[(result_df.lhs.str.contains('추천'))&(result_df.rhs.str.contains('여행지'))].sort_values(by=['lift'],ascending = False)
```
